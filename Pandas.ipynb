```python
# To check number of data points we are missing, you can simply do sum
# In this sum, True is counted as 1, and False is 0
titanic_df.isna().sum()
```

```python
# 'Name_len' column contains the length of name
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name', 'Name_len']].head()
```

```python
# 'Child_Adult' coulmn differtiate between child and adults by using if else
# IMPORTNAT! 'Child' should come before if (['Child' if x <=15] and then [else 'Adult'])
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x <=15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(10)
```

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else ('Adult' if x <=60 else 'Eldery'))
titanic_df['Age_cat'].value_counts()
```

```python
def get_category(age):
    cat = ''
    if age <=5: cat = 'Baby'
    elif age <=12: cat = 'Child'
    elif age <=18: cat = 'Teenager'
    elif age <=25: cat = 'Student'
    elif age <=35: cat = 'Young Adult'
    elif age <=60: cat = 'Adult'
    else : cat = 'Eldery'
    
    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
