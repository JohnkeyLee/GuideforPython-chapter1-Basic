```python
# To add another column named 'Age_0'
titanic_df['Age_0'] = 0
titanic_df.head()
```

```python
# This is how we creat the columns based on the exsiting data
titanic_df['Age_by_10'] = titanic_df['Age']*10
titanic_df['Family_No'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df.head()
```

```python
titanic_df['Age_by_10'] = titanic_df['Age_by_10'] + 100
titanic_df.head()
```

```python
# IMPORTANT!!
# DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
# 'labels', 'axis', and 'inplace' are key inputs
# axis=0 is indicating the rows and =1 is the columns in 2x3 DataFrame
# labels=1, axis=0 means Drop the second row. Regard 'labels' as 'index'ArithmeticError
titanic_drop_df = titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head()
```

```python
# In 'titanic_drop_df', there is no data for 'Age_0' because it is only for processing
titanic_df.head()
```

```python
# To drop in the original data set, we need to change 'inplace=False' to 'inplace=True'
drop_result = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
print('Output after changing to inplace=True:', drop_result)
titanic_df.head()
```

```python
# iloc[] is Location based indexing
# loc[] is Label based indexing
data = {'Name':['Jack', 'Jill', 'Wang', 'Mike'], 'Year':[2011, 2016, 2015, 2015], 'Gender':['Male', 'Female', 'Male', 'Male']}

data_df = pd.DataFrame(data, index=['one','two', 'three', 'four'])
print(data_df)
```

```python
data_df.iloc[0, 0]
# ERROR
# data_df.iloc[0, 'Name']
# data_df.iloc['one', 0]
```

```python
print(data_df.iloc[1,0])
print('\n',data_df.iloc[2,1])
print('\n',data_df.iloc[0:2, [0,1]])
print('\n',data_df.iloc[0:2, 0:3])
print('\n',data_df.iloc[:, [0,2]])
print('\n',data_df.iloc[:])
print('\n',data_df.iloc[:,:])
```

```python
print('\n',data_df.iloc[:,-1])
print('\n',data_df.iloc[:,:-1])
```
